***Loading Data***

```python
# Importing libraries for analysis

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("Clean Dataset.xlsx")
```

```python
data.info()
```

***Exploratory Data Analysis***

```python
data.head()
```

```python
data.columns
```

```python
years = data.columns[1:]
years
```

```python
# Add yearly average as a new column to the dataframe

data['Yearly Average'] = data.loc[:, years].mean(axis=1)
```

```python
data.head()
```

```python
# Showing the information about the data type

data.info()
```

```python
# Describing (overview) the numerical columns in our dataset 

print(data.describe())
```

```python
#Checking For Null values

data.isnull().sum()
```

***Performing Further EDA Using Chart***

```python
# Visualiazion of the total spending by country

plt.figure(figsize=(22, 6))
sns.barplot(data=data, x='Country Name', y='Yearly Average', order=data.sort_values('Yearly Average', ascending=False)['Country Name'])
plt.title('Average Educational Spending by Country')
plt.ylabel('Country')
plt.xlabel('Yearly Average')
plt.xticks(rotation=45)
plt.show()
```

```python
# Heatmap of spending across all years

plt.figure(figsize=(12,8))
sns.heatmap(data[years].corr(), cmap='coolwarm', annot=True)
plt.title('correlaton heatmap of educational spending across years')
plt.show()
```

```python
#univarariate analysis on the distribution of average spending across different countries

plt.figure(figsize=(10,6))
sns.histplot(data=data, x='Yearly Average', kde=True, bins=5)
plt.title('Distribution of Average Educational Spending ')
plt.xlabel('Average Spending')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(20, 6))
for country in data['Country Name'].unique():
    country_data = data[data['Country Name'] == country]
    plt.plot(country_data.columns[1:-1], country_data.iloc[0, 1:-1], label=country)
plt.title('Education Spending Over Time')
plt.xlabel('Year')
plt.ylabel('Percentage of GDP Spent on Education')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
for country in data['Country Name'].unique():
    country_data = data[data['Country Name'] == country]
    plt.figure(figsize=(20, 6))
    plt.plot(country_data.columns[1:-1], country_data.iloc[0, 1:-1], label=country)
    plt.title(f'Education Spending Over Time - {country}')
    plt.xlabel('Year')
    plt.ylabel('Percentage of GDP Spent on Education')
    plt.legend(loc='upper left')
    plt.show()
```
