```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("Dataset_for_timeseries.csv")
df.head()
```

```python
df.info()
```

```python
df["Country Name"].nunique()
```

```python
country_names = list(df.groupby("Country Name").groups.keys())
```

```python
# Dictionary to store DataFrames for each group
grouped_dfs = {}

grouped = df.groupby("Country Name")
# Iterate over groups
for group_name, group_data in grouped:
    grouped_dfs[group_name] = group_data.set_index(df.index[df['Country Name'] == group_name])
# Accessing DataFrames for each group
# print(grouped_dfs)

grouped_dfs[country_names[5]].reset_index()["percentage of GDP"].plot(kind="line")
```
